```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d-%m-%Y') #convert the reservation_status_date datatype from object to datetime
```

```python
df.drop(['company','agent'], axis=1, inplace = True) #to remove rows/cols. Mention th cols in a list that u want to remove. inplace= True to make changes to the data frame
df.dropna(inplace=True) #to remove the rows with null values
```

```python
df=df[df['adr']<5000]
```

```python
df=df[df['adr']>10]
```

```python
df=df[df['days_in_waiting_list']<366]
```

```python
df.describe()
```

```python
#Data Analysis and Visualisations
cancelled_percent = df['is_canceled'].value_counts(normalize=True)  #normalised = True returns the percentage, value_counts() returns number the no of times the value has occured
cancelled_percent
```

```python
print(cancelled_percent)

plt.figure(figsize =(5,5))
plt.title('Reservation status count')
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
ax1= sns.countplot(x='hotel', hue='is_canceled',data=df, palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel'] #filtering hotels on resort hotels
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']#filtering hotels on city hotels
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
resort_hotel
```

```python
city_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel
```

```python
plt.figure(figsize = (30,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
# Plot ADR for Resort Hotel with a specific color
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel', color='red')

# Plot ADR for City Hotel with a specific color
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel', color='orange')
plt.legend(fontsize=15)
plt.show()
```

```python
#to group the reservations by month
df['month']=df['reservation_status_date'].dt.month #extract the month from date and store in new column
plt.figure(figsize=(16,10))
ax2=sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')
legend_labels,_=ax2.get_legend_handles_labels()
ax2.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not_canceled', 'canceled'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('ADR Per Month', fontsize=30)
# Group by month, calculate the sum of 'adr' for canceled reservations, and reset the index
adr_per_month = df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()

# Use the barplot function with the corrected 'data' argument
sns.barplot(x='month', y='adr', data=adr_per_month)
plt.show()
```

```python
#To check top 10 countries with highest cancellation rate
cancelled_data = df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]#returns top 10 countries in descending order
plt.figure(figsize=(8,8))
plt.title("Top 10 countries with reservation canceled", fontsize=30)
plt.pie(top_10_country, autopct='%.2f', labels=top_10_country.index)
plt.show()
```

```python

cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace= True)

not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
plt.show()
```
